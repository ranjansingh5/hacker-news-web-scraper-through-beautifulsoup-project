Import Libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Scraping Logic

```python
news_info = []

for x in range(1,30):
    website = (f"https://news.ycombinator.com/?p={x}")

    r = requests.get(website)
    soup = BeautifulSoup(r.content, "html.parser")
    news = soup.find_all("tr", class_="athing submission")

    for item in news:
        title_tag = item.find("span", class_="titleline").find("a")
        title = title_tag.text.strip()
        news_link = title_tag["href"]

        rank_tag = item.find("span", class_="rank")
        rank = rank_tag.text.strip()

        all_news = {
            "Title": title,
            "Link": news_link,
            "Rank": rank
        }

        news_info.append(all_news)
```

Create DataFrame and Save to CSV

```python

df = pd.DataFrame(news_info)
df.to_csv("Hacker_News.csv", index=False)

print("Scraping Completed Successfully!")
```
